```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
% matplotlib inline
import seaborn as sns
sns.set_style('ticks')
palette = sns.color_palette("RdBu_r", n_colors=25)

from depynd import mutual_information
from depynd.mutual_information import mimat, cmimat
```

```python
iris = sklearn.datasets.load_iris()
df = pd.DataFrame(iris.data, columns=list(iris.feature_names))
df['target'] = iris.target
X = df.values
```

```python
mis_knn = mimat(X, method='knn')
cmis_knn = cmimat(X, method='knn')
```

```python
f, axes = plt.subplots(1, 2, figsize=[8, 4])
sns.heatmap(mis_knn, annot=True, center=0, cmap=palette, cbar=False, ax=axes[0])
sns.heatmap(cmis_knn, annot=True, center=0, cmap=palette, cbar=False, ax=axes[1])
```
