```python
import numpy as np
import pandas as pd
import sklearn.datasets
from scipy.special import digamma
import matplotlib.pyplot as plt
% matplotlib inline
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import seaborn as sns
sns.set_style('ticks')
palette = sns.color_palette("RdBu_r", n_colors=25)

from depynd import feature_selection
```

```python
iris = sklearn.datasets.load_iris()
df = pd.DataFrame(iris.data, columns=list(iris.feature_names))
df['target'] = iris.target
```

```python
selected = feature_selection.mifs(iris.data, iris.target)
```

```python
sns.pairplot(df, x_vars=np.array(iris.feature_names)[selected], y_vars='target')
```
