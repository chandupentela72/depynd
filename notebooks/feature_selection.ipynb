```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
% matplotlib inline
import seaborn as sns
sns.set_style('ticks')
palette = sns.color_palette("RdBu_r", n_colors=25)

from depynd.feature_selection import select
```

```python
iris = sklearn.datasets.load_iris()
df = pd.DataFrame(iris.data, columns=list(iris.feature_names))
df['target'] = iris.target
```

```python
selected_mrmr = select(iris.data, iris.target, lamb=0, method='mrmr')
selected_mifs = select(iris.data, iris.target, lamb=0, method='mifs')
```

```python
sns.pairplot(df, x_vars=np.array(iris.feature_names)[selected_mrmr], y_vars='target')
sns.pairplot(df, x_vars=np.array(iris.feature_names)[selected_mifs], y_vars='target')
```

```python
boston = sklearn.datasets.load_boston()
df = pd.DataFrame(boston.data, columns=list(boston.feature_names))
df['target'] = boston.target
```

```python
selected_mrmr = select(boston.data, boston.target, k=3, method='mrmr')
selected_mifs = select(boston.data, boston.target, k=3, method='mifs')
```

```python
sns.pairplot(df, x_vars=boston.feature_names[selected_mrmr], y_vars='target')
sns.pairplot(df, x_vars=boston.feature_names[selected_mifs], y_vars='target')
```
